# Heart Disease Prediction

```python
import numpy as np
import pandas as pd
```

```python
heart_df = pd.read_csv("https://raw.githubusercontent.com/tanmoyf2/Predicting-Heart-Disease/master/heart.csv")
heart_df.head(10)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('fivethirtyeight')
```

```python
#Basic Visualization
f, ax = plt.subplots(1, figsize=(16,12))
sns.heatmap(heart_df.corr(), annot=True, ax=ax,cmap="RdBu")
```

Chest Pain and Thalch have the highest effect on the outcome

```python
f, axes = plt.subplots(7,2, figsize=(14,25))
sns.violinplot(x=heart_df.target ,y=heart_df.age, ax=axes[0,0])
sns.violinplot(x=heart_df.target ,y=heart_df.sex, ax=axes[0,1])
sns.violinplot(x=heart_df.target ,y=heart_df.cp, ax=axes[1,0])
sns.violinplot(x=heart_df.target ,y=heart_df.trestbps, ax=axes[1,1])
sns.violinplot(x=heart_df.target ,y=heart_df.chol, ax=axes[2,0])
sns.violinplot(x=heart_df.target ,y=heart_df.fbs, ax=axes[2,1])
sns.violinplot(x=heart_df.target ,y=heart_df.restecg, ax=axes[3,0])
sns.violinplot(x=heart_df.target ,y=heart_df.thalach, ax=axes[3,1])
sns.violinplot(x=heart_df.target ,y=heart_df.exang, ax=axes[4,0])
sns.violinplot(x=heart_df.target ,y=heart_df.oldpeak, ax=axes[4,1])
sns.violinplot(x=heart_df.target ,y=heart_df.slope, ax=axes[5,0])
sns.violinplot(x=heart_df.target ,y=heart_df.ca, ax=axes[5,1])
sns.violinplot(x=heart_df.target ,y=heart_df.thal, ax=axes[6,0])
```

```python
column_names = heart_df.columns
column_names = column_names.drop('target')
for name in column_names:
    print('{}\n'.format(name))
    print(heart_df.groupby(['target'])[name].mean())
    print('*'*50)
    print()
```

In the dataset, people diagnosed with heart-disease had higher values of attributes (mean value) which are:cp,restecg,thalach and slope.
